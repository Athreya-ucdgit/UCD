```python
import numpy as np

np.random.seed(123)
```

```python
N = 20
d1 = np.random.rand(N) * (1 + pow(10,-5) * np.random.rand())
d2 = np.random.rand(N) * (1 + pow(10,-5) * np.random.rand())
d3 = np.random.rand(N) * (1 + pow(10,-5) * np.random.rand())
d4 = np.random.rand(N) * (1 + pow(10,-5) * np.random.rand())
d5 = d1 - d2 + d3 
```

```python
data = np.column_stack([d1,d2,d3,d4,d5])
data
```

```python
svd = np.linalg.svd(data)
U = svd.U
V_t = svd.Vh
S = np.diag(svd.S)

print(np.round(U,decimals=3),end="\n\n")
print(np.round(V_t,decimals=5),end="\n\n")
print(np.round(S,decimals=5),end="\n\n")
```

```python
z = np.zeros((np.shape(svd.U)[1]-np.shape(svd.S)[0],np.shape(svd.S)[0]))
S = np.vstack([S,z])
S[4,4] = 0
```

```python
result = U @ S @ V_t
result
```

```python
difference = data-result
frobenius_norm = np.linalg.norm(difference,"fro")
frobenius_norm
```

```python
for i in range(1,5):
    x = np.zeros((N,5))
    x[:i,:i] = S[:i,:i]
    result = U @ x @ V_t
    
    print("Frobenius Norm for Rank 1 approximation for r =",i,end=" ")
    print(np.linalg.norm(data - result,"fro"))

    print("Two Norm for Rank 1 approximation for r =",i,end=" ")
    print(np.linalg.norm(data - result,2),end="\n\n")
```
